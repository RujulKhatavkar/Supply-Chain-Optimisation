# Data Pre-Processing

```python
import pandas as pd
# Using cp1252 encoding
df = pd.read_csv("data/DataCoSupplyChainDataset.csv", encoding='cp1252')
pd.set_option('display.max_columns', None)

display(df.head())
```

## **Removing unnecessary columns**

```python
# List of columns to remove
cols_to_remove = [ 'Category Id', 'Customer Email','Customer Fname',
                  'Customer Lname','Customer Password','Department Id','Product Description',
                  'Product Image','Customer Street',]

# Remove the columns
df = df.drop(columns=cols_to_remove)

# Display the updated DataFrame
display(df.head())
```

```python
import numpy as np

# Set the random seed for reproducibility
np.random.seed(42)

# Calculate the number of refunds, returns, and missing orders
total_orders = len(df)
num_refunds = int(0.03 * total_orders)
num_returns = int(0.04 * total_orders)
num_missing = int(0.01 * total_orders)

# Create an array with the appropriate number of each category
post_order_action = (
    ['Refund'] * num_refunds +
    ['Return'] * num_returns +
    ['Missing'] * num_missing +
    ['None'] * (total_orders - num_refunds - num_returns - num_missing)
)

# Shuffle the array
np.random.shuffle(post_order_action)

# Add the array as a new column in the DataFrame
df['Post_Order_Action'] = post_order_action

# Print the value counts to verify the distribution
print(df['Post_Order_Action'].value_counts())
```

```python
# Define the percentage range for transportation cost
min_percentage = 0.05
max_percentage = 0.10

# Generate random percentages for each order within the specified range
random_percentages = np.random.uniform(min_percentage, max_percentage, size=len(df))

# Calculate the transportation cost for each order
df['Transportation_Cost'] = df['Order Item Total'] * random_percentages

# Print the first few rows to verify the new column
print(df[['Order Item Total', 'Transportation_Cost']].head())
```

## **Calculating Lead Time**

```python
# Rename columns (if needed)
df.rename(columns={
    'shipping date (DateOrders)': 'Shipping_Date',
    'order date (DateOrders)': 'Order_Date'
}, inplace=True)

# Convert to datetime
df['Shipping_Date'] = pd.to_datetime(df['Shipping_Date'])
df['Order_Date'] = pd.to_datetime(df['Order_Date'])

df['Lead_Time']=((df['Shipping_Date']-df['Order_Date']).dt.days)+df['Days for shipping (real)']
```

## **Creating Supplier IDs**

```python
# Define the list of possible Supplier IDs
supplier_ids = ['sup001', 'sup002', 'sup003', 'sup004', 'sup005']

# Define the probabilities for each Supplier ID
probabilities = [0.1, 0.2, 0.3, 0.25, 0.15]  # Sum should be 1

# Add a new column with weighted random Supplier IDs
df['Supplier_ID'] = np.random.choice(supplier_ids, size=len(df), p=probabilities)
```

```python
# Save the DataFrame to a new CSV file
df.to_csv("data.csv", index=False)
```
